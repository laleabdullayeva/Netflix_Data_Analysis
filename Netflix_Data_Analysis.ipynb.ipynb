```python
#For each type (Movie/TV Show), find the oldest and newest release year.
type_release_year=df.groupby('type')['release_year'].agg(['min','max'])
type_release_year
```

```python
#Create a bar chart showing the number of Movies vs TV Shows.
import matplotlib.pyplot as plt
type_counts=df['type'].value_counts()
type_counts.plot(kind='bar',color=['blue','green'])
plt.title('Number of Movie and Tv shows')
plt.xlabel('Type')
plt.ylabel('Number of titles')
plt.xticks(rotation=0)
plt.show()
```

```python
#  Create a pie chart of the distribution of ratings.
rating_counts=df['rating'].value_counts()
rating_counts
```

```python
rating_counts.plot(
                   kind='pie',
                   autopct='%1.1f%%',
                   startangle=90,
                   color=['red','blue','pink','green'],
                   figsize=(8,8)
                   )
plt.title('Rating count')
plt.ylabel('')
plt.show()
```

```python
titles_per_year = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(8,5))
titles_per_year.plot(kind='line', marker='o', color='purple')
plt.title('Number of Netflix Titles Released per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
#Show a horizontal bar chart of top 10 countries with most titles.
top_countries=df['country'].value_counts().head(10)
plt.figure(figsize=(8,6))
top_countries.plot(kind='barh',color='green',edgecolor='black')
plt.title('Top 10 countries')
plt.xlabel('Number of titles')
plt.ylabel('Country')
plt.show()
```

```python
#Plot a histogram of release_year distribution.
plt.figure(figsize=(8,5))
plt.hist(df['release_year'],bins=20,color='blue',edgecolor='black')
plt.title('release_year distribution')
plt.xlabel('Release year')
plt.ylabel('Number of titles')
plt.xticks(range(df['release_year'].min(), df['release_year'].max()+1))
plt.grid(axis='y', linestyle='--')
plt.show()
```
